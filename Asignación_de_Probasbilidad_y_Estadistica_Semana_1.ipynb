# **Conceptos Básicos**

- **Universo (U):**
El universo es el conjunto que contiene todos los elementos que nos interesan en un contexto dado.
U={x∣x es un elemento de interés}

- **Conjunto Vacío (∅):**
Es el conjunto que no contiene ningún elemento.
∅={}

- **Subconjunto (𝐴⊆𝐵):**
Decimos que 𝐴 es subconjunto de 𝐵 si todo elemento de 𝐴 pertenece a 𝐵.
A⊆Bsi∀x(x∈A⇒x∈B)

-** Intersección (𝐴∩𝐵):**
Es el conjunto formado por los elementos que están en 𝐴 y en 𝐵 al mismo tiempo.
A∩B={x∣x∈A y x∈B}

- **Unión (𝐴∪𝐵):**
Es el conjunto formado por todos los elementos que están en 𝐴 o en 𝐵 (o en ambos).
A∪B={x∣x∈A o x∈B}

- **Diferencia (𝐴−𝐵 o 𝐴∖𝐵):**
Es el conjunto de elementos que están en 𝐴 pero no en 𝐵.
A−B={x∣x∈A y x∈/B}

- **Complemento (𝐴𝑐  o 𝐴‾):**
Es el conjunto de todos los elementos del universo que no están en 𝐴.
Ac =U−A={x∈U∣x∈/A}

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles, venn3
import numpy as np
```

```python
# Definición de conjuntos
A = set([1, 2, 3, 4])
B = set([3, 4, 5, 6])

# Diagrama de Venn para dos conjuntos
plt.figure(figsize=(6,6))
v = venn2([A, B], set_labels=(r'$A$', r'$B$'))
plt.title(r'Intersección, Unión y Diferencia: $A$ y $B$')
plt.show()
```

```python
U = set(range(1, 11))  # Universo: números del 1 al 10
A = set([2, 4, 6, 8])

complemento_A = U - A
print("Universo U:", U)
print("Conjunto A:", A)
print("Complemento de A (A^c):", complemento_A)
```

```python
plt.figure(figsize=(6,6))
v = venn2(subsets=(len(A), len(U - A), 0), set_labels=(r'$A$', r'$A^c$'))
plt.title(r'Complemento de $A$ en $U$')
plt.show()
```
